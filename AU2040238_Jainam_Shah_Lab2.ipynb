```python
import random
import networkx as nx 
import matplotlib.pyplot as plt

def createEdgesTillConnect(graph, n):
    noOfEdges = 0

    while not nx.is_connected(graph):
        node1 = random.randint(1, n)
        node2 = random.randint(1, n)

        if not node1 == node2:
            noOfEdges += 1
            graph.add_edge(node1, node2)

    return noOfEdges 

def createNodes(graph, n):
    for i in range(1, n + 1):
        graph.add_node(i)

def generateGraphData(n):
    result = []
    result.append([])
    result.append([])

    tmp = n

    for i in range(1, 26):
        tmp += 10

        graph = nx.Graph()
        createNodes(graph, tmp)
        edgesRequired = createEdgesTillConnect(graph, tmp)

        result[0].append(tmp)
        result[1].append(edgesRequired)

    return result

def driverFunction():
    n = int(input())
    graphData = generateGraphData(n)

    print(graphData[1])
    plt.plot(graphData[0], graphData[1])
    plt.title("Number of edges required to make single connected component")
    plt.xlabel("Number of nodes")
    plt.ylabel("Number of edges required")

driverFunction()
```

```python
def calcAvgDegree(noOfEdges, n):
    return (2 * noOfEdges) / n

def createEdges(graph, n):
    noOfEdges = 0

    for i in range(1, n + 1):
        node1 = random.randint(1, n)
        node2 = random.randint(1, n)

        if not node1 == node2:
            noOfEdges += 1
            graph.add_edge(node1, node2)

    return noOfEdges

def generateDegreeHistogram(graph, n):
    result = []

    for i in range(1, n + 1):
        result.append(graph.degree[i])

    plt.hist(result)
    plt.xlabel("Degree")
    plt.ylabel("Number of nodes")
    plt.title("Degree Histogram")
    plt.show()

def secondDriverFunction():
    n = int(input())

    graph = nx.Graph()
    createNodes(graph, n)
    e = createEdges(graph, n)

    avgDegree = calcAvgDegree(e, n)

    print("Average Degree of Graph:", avgDegree)
    print("Clustering Coefficient of every node:", nx.clustering(graph))
    print("Average Clustering Value of Graph:", nx.average_clustering(graph))
    print("Local Bridges:", list(nx.local_bridges(graph)))

    generateDegreeHistogram(graph, n)

secondDriverFunction()
```
